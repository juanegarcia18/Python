```python
import pandas as pd
import numpy as np
from pandas import read_csv
from sklearn.preprocessing import LabelEncoder
import matplotlib
```

```python
pd.options.display.max_colwidth = 1000
pd.options.display.max_rows = 1000
```

```python
data_file = pd.read_csv("train.csv", nrows=1000)
data_file
```

```python
data_file.dtypes
```

```python
data_file.head(5)
```

```python
label_dict = dict()

for index in data_file.columns:
    if data_file[index].dtype == np.object:
        le = LabelEncoder()
  
        data_file[index] = le.fit_transform(data_file[index])
    
        label_dict[index] = dict()
        for cls, label in zip(le.classes_, le.transform(le.classes_)):
            label_dict[index][label] = cls
            
label_dict
```
