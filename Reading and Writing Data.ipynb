### Reading TXT Files into Parts or Partially

```python
pd.read_csv('Ch5 files\ch05_02.csv',skiprows=[2],nrows=3,header=None)
```

```python
#Another interesting and fairly common operation is to split into portions that part of the text on which you want to parse.
#Then, for each portion a specific operation may be carried out, in order to obtain an iteration, portion by portion.
#For example, you want to add the values in a column every three rows and then insert these sums in a series
out = pd.Series()
i = 0
pieces = pd.read_csv('Ch5 files\ch05_01.csv', chunksize=3)
for piece in pieces:
    #out.set_value(i, piece['white'].sum())
    out.at[i]=piece['white'].sum()
    i = i + 1
out
```
