```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
sns.scatterplot(x="tip", y="total_bill", data=tips,
                hue="day", size="size", palette="YlGnBu")
```

```python
sns.histplot(tips["tip"], kde=True, bins=15)
```

```python
sns.displot(tips["tip"], kde=True, bins=15)
```

```python
sns.barplot(x="sex", y="tip", data=tips, palette="YlGnBu")
```

```python
sns.boxplot(x="day", y="tip", data=tips, hue="sex", palette="YlGnBu")
```

```python
sns.stripplot(x="day", y="tip", data=tips, dodge=True, hue="sex", palette="YlGnBu")
```

```python
sns.jointplot(x="tip", y="total_bill", data=tips, cmap="YlGnBu", kind="kde", shade=True)
```

```python
tips.corr()
```

```python
sns.heatmap(tips.corr(),annot=True, cmap="coolwarm")
```
